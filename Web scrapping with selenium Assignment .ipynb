```python
#Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

import selenium

from selenium import webdriver 
import pandas as pd 
from selenium.webdriver.common.by import By 
import warnings 
warnings.filterwarnings('ignore')
import time
```

```python
driver = webdriver.Chrome()
driver.get("https://www.shine.com/")
```

```python
designation=driver.find_element(By.CLASS_NAME,"form-control")
designation.send_keys('Data Analyst')
```

```python
search=driver.find_element(By.CLASS_NAME,"searchForm_btnWrap_advance__VYBHN")
search.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]

title_tags=driver.find_elements(By.XPATH,'//h2[@itemprop="name"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)
    
    
location_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in location_tags[0:10]:
    location=i.text
    job_location.append(location)
    
company_tags=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in company_tags[0:10]:
    company=i.text
    company_name.append(company)
    
experience_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experience_tags[0:10]:
    exp=i.text
    experience_required.append(exp)
    
print(len(job_title),len(job_location),len(company_name),len(experience_required))

import pandas as pd 
df = pd.DataFrame({'Title':job_title,'Location':job_location,'Company_name':company_name,'Experience':experience_required})
df
```

```python
# Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:

import selenium

from selenium import webdriver 
import pandas as pd 
from selenium.webdriver.common.by import By 
import warnings 
warnings.filterwarnings('ignore')
import time

driver = webdriver.Chrome()
driver.get("https://www.flipkart.com")
```

```python
search_product=driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/header/div[1]/div[2]/form/div/div/input")
search_product.send_keys('sunglasses')
```

```python
search=driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/header/div[1]/div[2]/form/div/button")
search.click()
```

```python
sunglasses_data = []
total_sunglasses = 0

while total_sunglasses < 100:
  
    
    sunglasses_results = driver.find_elements(By.CLASS_NAME, "_1AtVbE")
    
    
    for sunglasses in sunglasses_results:
        try:
            brand = sunglasses.find_element(By.CLASS_NAME, "_2WkVRV").text
            description = sunglasses.find_element(By.CLASS_NAME, "IRpwTa").text
            price = sunglasses.find_element(By.CLASS_NAME, "_30jeq3").text
            
            sunglasses_data.append({"Brand": brand, "ProductDescription": description, "Price": price})
            total_sunglasses += 1
            
            if total_sunglasses >= 100:
                break
        except:
            
            continue
    
    
    try:
        next_button = driver.find_element(By.CLASS_NAME, "_1LKTO3")
        next_button.click()
    except:
        
        break
    
driver.quit()


df = pd.DataFrame(sunglasses_data)

print(df)
```

```python
#Q8: Write a python program to scrape data for Top 1000 Quotes of All Time

import selenium

from selenium import webdriver 
import pandas as pd 
from selenium.webdriver.common.by import By 
import warnings 
warnings.filterwarnings('ignore')
import time

driver = webdriver.Chrome()
driver.get("https://www.azquotes.com/")
```

```python
search_top=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
search_top.click()
```

```python
quotes_data = []
top_quotes = 0


while top_quotes < 1000:
   
    quote_elements = driver.find_elements(By.CLASS_NAME, "wrap-block")
    
    for quote_element in quote_elements:
        try:
            quote = quote_element.find_element(By.CLASS_NAME, "/html/body/div[1]/div[2]/div/div/div/div[1]/div/ul/li[1]/div/p/a[2]").text
            author = quote_element.find_element(By.CLASS_NAME, "/html/body/div[1]/div[2]/div/div/div/div[1]/div/ul/li[1]/div/div[1]/a").text
            quote_type = quote_element.find_element(By.CLASS_NAME, "/html/body/div[1]/div[2]/div/div/div/div[1]/div/ul/li[1]/div/div[2]/div[1]").text

            quotes_data.append({"Quote": quote, "Author": author, "TypeOfQuote": quote_type})
            top_quotes += 1

            if top_quotes >= 1000:
                break
        except:
            continue
    
    try:
        next_button = driver.find_element(By.XPATH, "/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]/a")
        next_button.click()
       
    except:
        break


driver.quit()


df = pd.DataFrame(quotes_data)

df
```
