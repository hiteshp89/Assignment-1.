```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException 
```

```python
driver = webdriver.Chrome()
driver.get('https://archive.ics.uci.edu/')
```

```python
accept = driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/div/div[2]/button")
accept.click()
```

```python
dataset = driver.find_element(By.XPATH,"//a[@class='btn-primary btn']")
dataset.click()
```

```python
data_urls =[]
start = 0
end = 64
for page in range(start,end):
    url = driver.find_elements(By.XPATH, '//a[@class="link-hover link text-xl font-semibold"]')
    for i in url:
        data_urls.append(i.get_attribute("href"))
    next_button = driver.find_element(By.XPATH, "/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]")
    if next_button.is_displayed():
        next_button.click()
        time.sleep(2)
    else:
        break
        
        
        
```

```python
        
data_list = []

for url in data_urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        # Locate and extract data from the current page
        dataset_name = driver.find_elements(By.XPATH, "//div[@class='relative flex flex w-full items-center gap-4 bg-primary p-2']/div[2]/div[1]/h1")
        data_type = driver.find_elements(By.XPATH, "//div[@class='grid grid-cols-8 gap-4 md:grid-cols-12']/div[1]/p")
        task = driver.find_elements(By.XPATH, "//div[@class='grid grid-cols-8 gap-4 md:grid-cols-12']/div[3]/p")
        attribute_type = driver.find_elements(By.XPATH, "//div[@class='grid grid-cols-8 gap-4 md:grid-cols-12']/div[4]/p")
        no_of_instances = driver.find_elements(By.XPATH, "//div[@class='grid grid-cols-8 gap-4 md:grid-cols-12']/div[5]/p")
        no_of_attribute = driver.find_elements(By.XPATH, "//div[@class='grid grid-cols-8 gap-4 md:grid-cols-12']/div[6]/p")
        year = driver.find_elements(By.XPATH, "//div[@class='relative flex flex w-full items-center gap-4 bg-primary p-2']/div[2]/h2")

        # Create a dictionary for the current page's data
        page_data = {
            'Dataset Name': '-',
            'Data Type': '-',
            'Task': '-',
            'Attribute Type': '-',
            'No. of Instances': '-',
            'No. of Attributes': '-',
            'Year': '-'
        }

        # Update the dictionary with actual data where available
        if dataset_name:
            page_data['Dataset Name'] = dataset_name[0].text
        if data_type:
            page_data['Data Type'] = data_type[0].text
        if task:
            page_data['Task'] = task[0].text
        if attribute_type:
            page_data['Attribute Type'] = attribute_type[0].text
        if no_of_instances:
            page_data['No. of Instances'] = no_of_instances[0].text
        if no_of_attribute:
            page_data['No. of Attributes'] = no_of_attribute[0].text
        if year:
            # Extract the year from the "Donated on" section
            year_text = year[0].text
            if "Donated on" in year_text:
                year_value = year_text.split("Donated on")[1].strip()
                page_data['Year'] = year_value

        # Append the dictionary to the list
        data_list.append(page_data)

    except NoSuchElementException:
        pass  # Handle the exception if needed, or just continue to the next page

# Create a DataFrame from the list of dictionaries
data = pd.DataFrame(data_list)        
```
